Import libraries

```python
import numpy as np
import pandas as pd
```

Read file


```python
df = pd.read_csv('/content/drive/MyDrive/Iris.csv')
```

```python
#divide test data and train data
train_df = df.sample(frac=.8);
test_df = df.sample(frac=.2);
train_df.reset_index(drop=True, inplace=True)
test_df.reset_index(drop=True, inplace=True)
```

```python
test_df
```

Main function

```python
def KNN(k):
    for i in range(len(test_df)):
        s_list = []
        for j in range(len(train_df)):
            eq = (test_df['SepalLengthCm'][i] - train_df['SepalLengthCm'][j])**2 + \
                 (test_df['SepalWidthCm'][i] - train_df['SepalWidthCm'][j])**2 + \
                 (test_df['PetalLengthCm'][i] - train_df['PetalLengthCm'][j])**2 + \
                 (test_df['PetalWidthCm'][i] - train_df['PetalWidthCm'][j])**2

            s_list.append([eq, train_df['Species'][j]])

        s_list.sort(key=lambda x: x[0])
        top_k = s_list[:k]

        s = 0
        vg = 0
        vs = 0
        for n in top_k:
            if n[1] == "Iris-setosa":
                s += 1
            elif n[1] == "Iris-virginica":
                vg += 1
            elif n[1] == "Iris-versicolor":
                vs += 1



        if s > vg and s > vs:
            test_df.at[i, f"validate{k}"] = "Iris-setosa"
        elif vg > s and vg > vs:
            test_df.at[i, f"validate{k}"] = "Iris-virginica"
        else:
            test_df.at[i, f"validate{k}"] = "Iris-versicolor"

    return 0



test_df
```

```python
KNN_r3 = KNN(3)
KNN_r5 = KNN(5)
KNN_r7 = KNN(7)


for j in range(3,8,2):
  validation = 0
  for i in range(len(test_df)):
    if(test_df['Species'][i] == test_df[f'validate{j}'][i]):
      validation+=1
  print(f"The accuracy for {j} =" , validation/len(test_df)*100,"%")


```

```python
cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
norm_train_df = pd.DataFrame()
norm_train_df[cols] = df[cols]
norm_test_df = pd.DataFrame()
norm_test_df[cols] = df[cols]



for col in cols:
    mean = train_df[col].mean()
    std = train_df[col].std()
    norm_train_df[col] = (train_df[col] - mean) / std
    mean = test_df[col].mean()
    std = test_df[col].std()
    norm_test_df[col] = (test_df[col] - mean) / std
norm_test_df['Species'] = test_df["Species"]
norm_train_df['Species'] = train_df["Species"]

norm_test_df.dropna()
```

```python
def KNNNormalized(k):
    for i in range(len(norm_test_df)):
        s_list = []
        for j in range(len(norm_train_df)):
            eq = (norm_test_df['SepalLengthCm'][i] - norm_train_df['SepalLengthCm'][j])**2 + \
                 (norm_test_df['SepalWidthCm'][i] - norm_train_df['SepalWidthCm'][j])**2 + \
                 (norm_test_df['PetalLengthCm'][i] - norm_train_df['PetalLengthCm'][j])**2 + \
                 (norm_test_df['PetalWidthCm'][i] - norm_train_df['PetalWidthCm'][j])**2
            s_list.append([eq, norm_train_df['Species'][j]])

        s_list.sort(key=lambda x: x[0])
        top_k = s_list[:k]

        s = 0
        vg = 0
        vs = 0
        for n in top_k:
            if n[1] == "Iris-setosa":
                s += 1
            elif n[1] == "Iris-virginica":
                vg += 1
            elif n[1] == "Iris-versicolor":
                vs += 1


        if s > vg and s > vs:
            norm_test_df.at[i, f"validate{k}"] = "Iris-setosa"
        elif vg > s and vg > vs:
            norm_test_df.at[i, f"validate{k}"] = "Iris-virginica"
        else:
            norm_test_df.at[i, f"validate{k}"] = "Iris-versicolor"

    return 0



norm_test_df.dropna()
```
